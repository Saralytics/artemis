```python
import os
import json
```

```python
auth_str = 'curl -X POST "https://accounts.spotify.com/api/token" \
     -H "Content-Type: application/x-www-form-urlencoded" \
     -d "grant_type=client_credentials&client_id=41c881206da74f958c72a86e2c534a4a&client_secret=dd7c33b58f194e36b1fc378f7dc5fad4"'

# auth_str = 'curl -X POST "https://accounts.spotify.com/api/token" \
#      -H "Content-Type: application/x-www-form-urlencoded" \
#      -d "grant_type=client_credentials&client_id=41c881206da74f958c72a86e2c534a4a"'
```

```python
def get_auth_token(auth_str):
    """
    :auth_str: the post request to get back access token. 
    reference: https://developer.spotify.com/documentation/web-api/tutorials/getting-started
    """
    try:
        res = os.popen(auth_str).read()
        res = json.loads(res)
        # if res['error']:
        #     print('There is an error: ' + res['error'])
        #     return
        # else:
        #     return res['access_token']
        return res['access_token']
        
    except Exception:
        raise Exception
```

```python
token = get_auth_token(auth_str)
token
```
